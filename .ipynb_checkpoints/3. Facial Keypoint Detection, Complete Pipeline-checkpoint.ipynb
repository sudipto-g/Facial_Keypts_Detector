## Face and Facial Keypoint detection

After a neural network has been trained to detect facial keypoints, the network is applied to *any* image that includes faces.  
The neural network expects a Tensor of a certain size as input and, so, to detect any face, some pre-processing needs to be done.

1. Detection of all the faces in an image using a face detector (using a Haar Cascade detector in this notebook).
2. Pre-processing of the face images so that they are grayscale, and transformed to a Tensor of the input size that the network expects. This step will be similar to the `data_transform` created and applied in Notebook 2, whose job was to rescale, normalize, and turn any image into a Tensor to be accepted as input to the defined CNN.
3. Use the trained model to detect facial keypoints on the image.

---

Loading in required libraries for this section of the project.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

#### Select an image 

Select an image to perform facial keypoint detection on; you can select any image of faces in the `images/` directory.

```python
import cv2
# loading in color image for face detection
image = cv2.imread('images/obamas.jpg')

# switching red and blue color channels 
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# ploting the image
fig = plt.figure(figsize=(9,9))
plt.imshow(image)
```

## Detect all faces in an image

Next, one of OpenCV's pre-trained Haar Cascade classifiers will be used, all of which can be found in the `detector_architectures/` directory, to find any faces in the selected image.

In the code below, each face in the original image is looped over and a red square is drawn on each face (in a copy of the original image, so as not to modify the original). 

An example of face detection on a variety of images is shown below.

<img src='images/haar_cascade_ex.png' width=80% height=80%/>


```python
# loading in a haar cascade classifier for detecting frontal faces
face_cascade = cv2.CascadeClassifier('detector_architectures/haarcascade_frontalface_default.xml')

# running the detector
# the output here is an array of detections; the corners of each detection box
faces = face_cascade.detectMultiScale(image, 1.2, 2)

# making a copy of the original image to plot detections on
image_with_detections = image.copy()

# looping over the detected faces, and marking the image where each face is found
for (x,y,w,h) in faces:
    # drawing a rectangle around each detected face
    # the width of the rectangle may need to be changed depending on image resolution
    cv2.rectangle(image_with_detections,(x,y),(x+w,y+h),(255,0,0),3) 

fig = plt.figure(figsize=(9,9))

plt.imshow(image_with_detections)
```

## Loading in a trained model

Once an image is obtained to work with (and, again, any image of faces in the `images/` directory can be chosen), the next step is to pre-process that image and feed it into the trained CNN facial keypoint detector.

Loading in the trained model by its filename.

```python
import torch
from models import Net

net = Net()

net.load_state_dict(torch.load('keypoints_model.pt'))

# printing out the network and preparing it for testing
net.eval()
```

## Keypoint detection

Now, each detected face in an image will be looped over, and the faces will be transformed into Tensors that the trained CNN can accept as input images.

### To transform each detected face into an input Tensor

1. Convert the face from RGB to grayscale
2. Normalize the grayscale image so that its color range falls in [0,1] instead of [0,255]
3. Rescale the detected face to be the expected square size for the trained CNN (96x96, in my case)
4. Reshape the numpy image into a torch image.


### To detect and display the predicted keypoints

After each face has been appropriately converted into an input Tensor for the network to see as input, the tensor will be wrapped in a Variable() and the `net` can be applied to each face.  
The ouput should be the predicted the facial keypoints. These keypoints will need to be "un-normalized" for display.  
The final image may look like the following with facial keypoints that closely match the facial features on each individual face:

<img src='images/michelle_detected.png' width=30% height=30%/>




```python
def show_all_keypoints(image, predicted_key_pts):
    """To show image with predicted keypoints"""
    # image is grayscale
    plt.figure()
    plt.imshow(image, cmap='gray')
    plt.scatter(predicted_key_pts[:, 0], predicted_key_pts[:, 1], s=5, marker='.', c='m')
```

```python
image_copy = np.copy(image)

# loop over the detected faces from your haar cascade
for (x,y,w,h) in faces:
    
    # Select the region of interest that is the face in the image 
    roi = image_copy[y-50:y+h+50, x-40:x+w+40] # the numeric values are needed for scaling the output keypoints correctly.
    width_roi = roi.shape[1] # needed later for scaling keypoints
    height_roi = roi.shape[0] # needed later for scaling keypoints
    roi_copy = np.copy(roi) # will be used as background to display final keypints.
    
    ## TODO: Convert the face region from RGB to grayscale
    roi = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)

    ## TODO: Normalize the grayscale image so that its color range falls in [0,1] instead of [0,255]
    roi = roi/255.0
    
    ## TODO: Rescale the detected face to be the expected square size for your CNN (224x224, suggested)
    roi = cv2.resize(roi, (96, 96))  # resize the image to get a square 96*96 image.
    roi = np.reshape(roi,(96, 96, 1)) # reshape after rescaling to add the third color dimension.
    
    ## TODO: Reshape the numpy image shape (H x W x C) into a torch image shape (C x H x W)
    roi = roi.transpose(2, 0, 1)
    
    ## TODO: Make facial keypoint predictions using your loaded, trained network 
    roi = torch.from_numpy(roi).type(torch.FloatTensor) # convert images to FloatTensors (common source of error)
    
    # Runtime error : expected stride to be a single integer value or a list of 1 values to match the convolution dimensions, 
    # but got stride=[1, 1]. This error will occur if the following line is omitted. Hence we need to apply unsqueeze operation.
    roi = roi.unsqueeze(0)
    
    # Pass the transformed input images to the network for detecting keypoints.
    keypoints = net(roi)
    keypoints = keypoints.view(68, 2)
    
    # Undo the transformations performed on the facial keypoints
    keypoints = keypoints.data.numpy()
    keypoints = keypoints*50.0 + 100
    
    ## TODO: Display each detected face and the corresponding keypoints        
    keypoints = keypoints * (width_roi / 96, height_roi / 90) # scale the keypoints to match the size of the output display.
    plt.figure() 
     
    # Using helper function for display as defined previously.  
    show_all_keypoints(roi_copy, keypoints)
    plt.show()
```
